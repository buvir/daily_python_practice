# Missing Value handling

Importance of Handling Missing Values in Data Analysis
Missing values (or null values) in a dataset refer to the absence of data in certain observations. Handling missing values is crucial in data preprocessing because they can significantly impact the quality and reliability of analysis, machine learning models, and decision-making.

Why Handling Missing Values is Important?
1. Ensures Data Accuracy & Completeness
Missing values can distort statistical measures like mean, median, and standard deviation.

Incomplete data may lead to incorrect conclusions in business or research.

2. Improves Machine Learning Model Performance
Many ML algorithms (e.g., regression, neural networks) cannot handle missing values and may produce errors.

Missing data can bias model training, leading to poor predictions.

3. Avoids Biased Analysis
If missing values are not random (e.g., certain groups not responding in a survey), ignoring them can introduce bias.

Example: In medical data, missing patient records may hide critical trends.

4. Maintains Data Consistency
Some datasets require complete records for time-series or relational analysis.

Missing values can break joins in SQL or cause misalignment in time-based data.

5. Enhances Data Visualization
Graphs and charts may misrepresent trends if missing values are not handled properly.

Common Methods to Handle Missing Values
Method	Description	When to Use
Deletion (Dropping rows/columns)	Remove missing entries.	When missing data is small and random.
Mean/Median/Mode Imputation	Replace missing values with average, median, or most frequent value.	For numerical/categorical data with low variability.
Forward/Backward Fill	Fill missing values with the previous or next value.	Time-series data with ordered observations.
Interpolation	Estimate missing values using mathematical methods (linear, polynomial).	Time-series or sequential data.
Predictive Modeling	Use ML models (KNN, regression) to predict missing values.	When data has strong correlations.
Assigning a New Category (e.g., "Unknown")	Treat missing values as a separate category.	For categorical data with many missing entries.

### **Missing Data Mechanisms**  

| Type | Description | Handling Approach | Python Example |
|------|------------|------------------|---------------|
| **MCAR**<br>(Missing Completely at Random) | Missingness is random, no pattern. | Safe to delete or impute. | `df.dropna()` or `df.fillna(df.mean())` |
| **MAR**<br>(Missing at Random) | Missingness depends on observed data. | Impute using related variables. | `df['Age'].fillna(df.groupby('Education')['Age'].mean(), inplace=True)` |
| **MNAR**<br>(Missing Not at Random) | Missingness depends on unobserved data. | Requires domain knowledge/modeling. | `# Advanced: Use ML models (e.g., KNNImputer)` |  

**Key Notes:**  
- **MCAR:** No bias → Simple deletion/mean imputation.  
- **MAR:** Leverage correlated features for imputation.  
- **MNAR:** Proceed with caution (potential bias).  

#When to Use Which?
**Mean/Median**: Numerical data, small missingness

**Mode:** Categorical data

**Arbitrary Value:** When missingness has meaning

**Forward/Backward Fill**: Time-series data

**Iterative Imputer:** High correlation between features

**KNN Imputer:** Similar patterns exist in data

## Univariate Imputation Methods

### 1. Mean Imputation
**When to use:** Numerical data with normal distribution  
**Pros:** Simple, preserves mean  
**Cons:** Underestimates variance  

```python

When to use: Numerical data with outliers
Pros: Robust to outliers

#1. Sample Dataset (Before Cleaning)

````python
import pandas as pd
import numpy as np

data = {
    "Age": [25, 30, 22, 200, 28, np.nan, 35, 19, 150, 27],  # Outliers: 200, 150
    "Income ($)": [50000, 55000, 48000, 5000000, 52000, np.nan, 60000, 45000, 3000000, 51000],  # Outliers: 5M, 3M
    "Gender": ["M", "F", "M", np.nan, "F", "F", "M", "M", np.nan, "F"]  # Missing values
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)
````

2. Handling Missing Values
A. Mean Imputation (for "Age")

````python
from sklearn.impute import SimpleImputer

mean_imputer = SimpleImputer(strategy="mean")
df["Age_MeanImputed"] = mean_imputer.fit_transform(df[["Age"]])
print("\nAfter Mean Imputation (Age):\n")
print(df[["Age", "Age_MeanImputed"]])
````

B. Median Imputation (Better for "Age")

````python
median_imputer = SimpleImputer(strategy="median")
df["Age_MedianImputed"] = median_imputer.fit_transform(df[["Age"]])
print("\nAfter Median Imputation (Age):\n")
print(df[["Age", "Age_MedianImputed"]])
````

C. Mode Imputation (for "Gender")

````python
mode_imputer = SimpleImputer(strategy="most_frequent")
# Flatten the output to 1D array using .ravel() or select first column
df["Gender_ModeImputed"] = mode_imputer.fit_transform(df[["Gender"]]).ravel()
# Alternative:
# df["Gender_ModeImputed"] = mode_imputer.fit_transform(df[["Gender"]])[:, 0]

print("\nAfter Mode Imputation (Gender):\n")
print(df[["Gender", "Gender_ModeImputed"]])
````

3. Handling Outliers
A. Detect Outliers (IQR Method)

````python
Q1 = df["Income ($)"].quantile(0.25)
Q3 = df["Income ($)"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Income ($)"] < lower_bound) | (df["Income ($)"] > upper_bound)]
print("\nOutliers in Income:")
print(outliers[["Age", "Income ($)"]])
````

B. Capping Outliers (Winsorization)

````python
df["Income_Capped"] = df["Income ($)"].clip(lower=lower_bound, upper=upper_bound)
print("\nAfter Capping Outliers (Income):\n")
print(df[["Income ($)", "Income_Capped"]])
````

 5M & 3M replaced with 122,500 (upper bound).

4. Final Cleaned Data

````python
final_df = df.drop(["Age", "Income ($)", "Gender"], axis=1)
print("\nFinal Cleaned Data:\n")
print(final_df)
````

1. Python Code for IQR & Outliers

````python
import numpy as np
import matplotlib.pyplot as plt

# Dataset (example: exam scores)
data = np.array([55, 60, 65, 70, 72, 75, 76, 78, 80, 85, 95, 98, 30, 100])

# Calculate Q1, Q3, IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data < lower_bound) | (data > upper_bound)]

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower bound (Q1 - 1.5*IQR): {lower_bound}")
print(f"Upper bound (Q3 + 1.5*IQR): {upper_bound}")
print(f"Outliers: {outliers}")
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Dataset (exam scores with outliers)
data = np.array([55, 60, 65, 70, 72, 75, 76, 78, 80, 85, 95, 98, 30, 100])

# Calculate IQR bounds
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Separate outliers and normal data
outliers = data[(data < lower_bound) | (data > upper_bound)]
normal_data = data[(data >= lower_bound) & (data <= upper_bound)]

# Plot
plt.figure(figsize=(8, 6))
box = plt.boxplot(data, vert=True, patch_artist=True, showfliers=False)  # Hide default outliers

# Customize box colors
box['boxes'][0].set_facecolor('lightblue')
box['medians'][0].set_color('red')

# Add jittered normal data points (avoid overlap)
x_jitter = np.random.normal(1, 0.05, size=len(normal_data))
plt.scatter(x_jitter, normal_data, color='green', alpha=0.5, label='Normal Data')

# Add outliers
x_outliers = np.random.normal(1, 0.05, size=len(outliers))
plt.scatter(x_outliers, outliers, color='red', marker='X', s=100, label='Outliers')

# Annotations
plt.title("Boxplot with Overlaid Data Points", fontsize=14)
plt.xlabel("Distribution", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

IQR Outlier Detection Code

````python
import numpy as np

def detect_outliers_iqr(data):
    """
    Detect outliers using the IQR method.
    Returns: Normal data, outliers, and bounds.
    """
    # Calculate quartiles and IQR
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    normal_data = data[(data >= lower_bound) & (data <= upper_bound)]

    return {
        "normal_data": normal_data,
        "outliers": outliers,
        "bounds": {"lower": lower_bound, "upper": upper_bound},
        "IQR": IQR,
        "quartiles": {"Q1": Q1, "Q3": Q3}
    }

# Example usage
data = np.array([55, 60, 65, 70, 72, 75, 76, 78, 80, 85, 95, 98, 30, 100])
result = detect_outliers_iqr(data)

print("Normal Data:", result["normal_data"])
print("Outliers:", result["outliers"])
print("Lower Bound (Q1 - 1.5*IQR):", result["bounds"]["lower"])
print("Upper Bound (Q3 + 1.5*IQR):", result["bounds"]["upper"])
print("IQR:", result["IQR"])
print("Quartiles (Q1, Q3):", result["quartiles"])
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.boxplot(data, vert=True, patch_artist=True, showfliers=False)  # Hide default outliers
plt.scatter([1] * len(result["normal_data"]), result["normal_data"], color="green", label="Normal Data")
plt.scatter([1] * len(result["outliers"]), result["outliers"], color="red", marker="X", label="Outliers")
plt.title("IQR Outlier Detection")
plt.ylabel("Values")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
````

#Z-score

z=raw score- mean/standard deviation

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def detect_outliers_zscore(data, threshold=3):
    """
    Detect outliers using the Z-score method.
    Returns: Normal data, outliers, and Z-scores.
    """
    # Calculate Z-scores
    z_scores = np.abs(stats.zscore(data))

    # Identify outliers
    outliers = data[z_scores > threshold]
    normal_data = data[z_scores <= threshold]

    return {
        "normal_data": normal_data,
        "outliers": outliers,
        "z_scores": z_scores,
        "threshold": threshold
    }

# Example dataset (same as IQR example for comparison)
data = np.array([55, 60, 65, 70, 72, 75, 76, 78, 80, 85, 95, 98, 30, 100])

# Detect outliers
z_result = detect_outliers_zscore(data)
iqr_result = detect_outliers_iqr(data)  # From previous IQR function

print("=== Z-Score Results ===")
print("Normal Data:", z_result["normal_data"])
print("Outliers:", z_result["outliers"])
print("Z-Scores:", z_result["z_scores"].round(2))
print("Threshold:", z_result["threshold"])

print("\n=== IQR Results ===")
print("Outliers:", iqr_result["outliers"])
````

Key Differences: Z-Score vs. IQR
Feature	Z-Score Method	IQR Method
Assumption	Data is normally distributed	No distribution assumptions
Sensitivity	Affected by extreme values	Robust to extreme values
Threshold	Typically ±2.5 or ±3 std	Q1 - 1.5IQR / Q3 + 1.5IQR
Best For	Gaussian data	Skewed distributions

````python
plt.figure(figsize=(12, 5))

# Z-Score Plot
plt.subplot(1, 2, 1)
plt.scatter(range(len(data)), data,
            c=np.abs(stats.zscore(data)) > 3,
            cmap='cool', s=100)
plt.axhline(y=np.mean(data), color='green', linestyle='--', label='Mean')
plt.title("Z-Score Outlier Detection")
plt.xlabel("Data Index")
plt.ylabel("Values")
plt.legend()

# IQR Plot
plt.subplot(1, 2, 2)
box = plt.boxplot(data, patch_artist=True, showfliers=False)
plt.scatter([1]*len(data), data,
            c=(data < iqr_result["bounds"]["lower"]) | (data > iqr_result["bounds"]["upper"]),
            cmap='cool', s=100)
plt.title("IQR Outlier Detection")
plt.suptitle("Comparison of Outlier Detection Methods", y=1.05)
plt.tight_layout()
plt.show()
````

# Data Visualization Cheat Sheet

| Chart Type          | Description                                                                 | Best Use Case                                                                 |
|---------------------|-----------------------------------------------------------------------------|-------------------------------------------------------------------------------|
| **Pie Chart**       | Circular chart divided into slices to show proportional composition         | Showing market share or simple part-to-whole relationships (3-6 categories)   |
| **Donut Chart**     | Variation of pie chart with center cut out, allowing additional information | Survey results with neutral options or when center space can display metrics  |
| **Stacked Column**  | Columns divided into colored segments showing sub-category contributions    | Monthly revenue by product category (showing total + composition)             |
| **Treemap**         | Nested rectangles sized by value, showing hierarchical part-to-whole        | File system storage allocation or organizational budget breakdown            |
| **Line Chart**      | Points connected by lines showing values over time                          | Tracking monthly active users or stock prices over time                       |
| **Multi-Line Chart**| Multiple lines on same axes for comparison                                 | Comparing stock prices of 3-5 companies over time                             |
| **Stacked Area**    | Filled areas showing cumulative totals and composition over time            | Website traffic sources by month (shows growth + source changes)              |
| **Spline Chart**    | Smoothed line chart with curved interpolation between points               | Displaying smoothed trends in noisy data like sensor readings                 |
| **Histogram**       | Bars showing frequency distribution of continuous data                      | Visualizing employee age distribution or call duration frequency              |
| **Box Plot**        | Shows quartiles, median and outliers of distribution                        | Comparing test scores across different classes or groups                      |
| **Violin Plot**     | Combines box plot with rotated kernel density plot                          | Showing salary distribution by department (density + statistics)              |
| **Density Plot**    | Smoothed curve showing probability distribution                             | Analyzing probability distribution of continuous variables like income        |
| **Bar Chart**       | Horizontal bars comparing discrete categories                               | Comparing quarterly sales across regions (especially with long category names)|
| **Column Chart**    | Vertical bars comparing discrete categories                                 | Monthly revenue comparisons (time-series categories work well vertically)     |
| **Scatter Plot**    | Points showing relationship between two continuous variables                | Revealing correlation between height and weight across a population           |
| **Connected Scatter**| Scatter plot with points connected in sequence                             | Life expectancy vs. GDP evolution over time (shows trajectory)               |
| **Bubble Chart**    | Scatter plot with point size representing third variable                    | Country comparisons (GDP on x, HDI on y, population as bubble size)          |
| **Word Cloud**      | Words sized by frequency in text data                                       | Visualizing prominent themes in customer feedback or document keywords        |

# Chart Selection by Data Type

## Numerical Data
### Continuous (age, weight, temperature)
- **Distribution**: Histogram, Density plot, Box plot, Violin plot
- **Trends**: Line chart, Area chart
- **Relationships**: Scatter plot, Bubble chart

### Discrete (counts, whole numbers)
- **Comparison**: Bar chart, Column chart
- **Composition**: Stacked bar/column, Pie/Donut (for few categories)
- **Distribution**: Histogram

## Categorical Data
### Nominal (no order: eye color, blood type)
- **Comparison**: Bar chart, Column chart
- **Composition**: Pie chart, Donut chart, Treemap
- **Word Data**: Word cloud

### Ordinal (ordered categories: rating scales, education level)
- **Comparison**: Ordered bar/column chart
- **Distribution**: Histogram
- **Trends**: Line chart (if time-based)

## Mixed Data Types
- **Numerical vs Categorical**: Box plot, Violin plot
- **Two Numerical + One Categorical**: Grouped scatter plot, Faceted charts
- **Time + Numerical**: Line chart, Area chart
- **Time + Categorical**: Stacked area/column chart

## Special Cases
- **Hierarchical Data**: Treemap, Sunburst
- **Geospatial Data**: Choropleth, Bubble maps
- **Text Data**: Word cloud

Stacked Column

2. Trend Charts

Line Chart

````python
months = pd.date_range(start='2023-01', periods=12, freq='M')
sales = [100, 120, 90, 150, 200, 180, 220, 250, 230, 300, 280, 320]
plt.plot(months, sales)
plt.title('Monthly Sales Growth')
````

Spline Chart
